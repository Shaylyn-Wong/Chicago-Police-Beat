```python
df1=chicargo_crime_df
```

```python
df2 = df1.dropna()
df2.head()
```

```python
df2.shape[0]
```

```python
df3=df2[['Beat','Year']]
df3.head(5)
```

```python
print(df3)
```

```python
crime_df=df3.groupby(["Beat"])["Beat"].count().reset_index(name="count")
crime_df.head(5)
```

```python
map_df=crime_df.set_index('Beat').join(Beat_df.set_index('BEAT_NUM'))
map_df
```

```python
map_gdf = gpd.GeoDataFrame(map_df, geometry='geometry')
```

```python
print(type(map_gdf))
```

```python
map_gdf.info()
```

```python
map_gdf.plot()
map_gdf["center"] = map_gdf["geometry"].centroid
```

```python
import matplotlib.pyplot as plt
```

```python
variable ='count'
vmin, vmax = 20, 2000
fig, ax = plt.subplots(1, figsize=(12,12))
map_gdf.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')
ax.axis('off')
ax.set_title('Chicargo Crimes 2020', fontdict={'fontsize': '25', 'fontweight' :'3'})
ax.annotate('Source: Chicago Data Portal, 2021',xy=(0.1, .08),  xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12, color='#555555')
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)
```

```python
fig.savefig('crime_map_2020.png', dpi=300)
```
